```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly

# data = pd.read_csv("/kaggle/input/netflix-stock-price-v2/Netflix Stock Data.csv")
data = pd.read_csv("/kaggle/input/nflx-5/NFLX (5).csv")

print("Original Data: ")
print(data.head())

# we only need the date and close columns
data = data[["Date", "Close"]]
print("Data only has date and close columns:")
print(data.head())

# prophet needs the columns to be named 'ds' and 'y'
data.columns = ["ds", "y"]

# uses all past data from stock
prophet_long = Prophet(daily_seasonality=True)
prophet_long.fit(data)
future_dates_long = prophet_long.make_future_dataframe(periods=(len(data) + 100)) # predicts up to 100 days in the future
predictions_long = prophet_long.predict(future_dates_long)

# uses only data from past month
PM_data = data.tail(30) # PM - past month
prophet_short = Prophet(daily_seasonality=True)
prophet_short.fit(PM_data)
print(len(PM_data))
future_dates_short = prophet_short.make_future_dataframe(periods=(len(PM_data) + 10)) # predicts up to 10 days in the future
predictions_short = prophet_short.predict(future_dates_short)


print("All time stock graph: ")
plot_plotly(prophet_long, predictions_long)

print("Past month stock graph: ")
plot_plotly(prophet_short, predictions_short)

print(predictions_short.head())
print(predictions_long.head())

print("Prediction: ")

for i in range(25, 30):
    print(predictions_long['yhat'][len(data) - i] - 0.0 * predictions_short['yhat'][len(PM_data) - i])

# print((1 * predictions_long['yhat'][len(data)] + 0 * predictions_short['yhat'][len(PM_data)]))
```

```python
# use yahoo api --> predict 1 day with all time(30%) and past month(70%)
# goal --> 10 hot stocks with automated calculation every night
import pandas as pd

data = pd.read_csv("/kaggle/input/netflix-stock-price-v2/Netflix Stock Data.csv")
```

```python
from prophet import Prophet
```

```python
prophet = Prophet(daily_seasonality=True)
prophet.fit(data)
```

```python
future_dates = prophet.make_future_dataframe(periods=1200)
predictions = prophet.predict(future_dates)
```

```python
from prophet.plot import plot_plotly
from prophet import plot

plot_plotly(prophet, predictions)
```

```python
unknown_data = data.iloc[-180:]
data = data.iloc[:-180]
```

```python
prophet = Prophet(daily_seasonality=True)
prophet.fit(data)

future_dates = prophet.make_future_dataframe(periods=365)
predictions = prophet.predict(future_dates)
```

```python
plot_plotly(prophet, predictions)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

pred = predictions[predictions['ds'].isin(unknown_data['ds'])]

plt.plot(pd.to_datetime(unknown_data['ds']), unknown_data['y'], label="Actual")
plt.plot(pd.to_datetime(unknown_data['ds']), pred['yhat'], label="Prediction")

plt.legend()
```
